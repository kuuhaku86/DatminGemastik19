```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filepath = "data.xls"
electricity_data = pd.read_excel(filepath)
```

```python
electricity_data.head()
```

```python
electricity_data.shape
```

```python
electricity_data.columns
```

```python
plt.figure(figsize=(20,6))
sns.countplot(electricity_data['IncomeGroup'])
```

```python
plt.figure(figsize=(20,6))
sns.countplot(electricity_data['Region'])
```

```python
electricity_data.describe()
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN',strategy='mean',axis=0)
numerical_features = electricity_data.select_dtypes(include='float64')
numerical_features = numerical_features.columns
imputer = imputer.fit(electricity_data[numerical_features])
electricity_data[numerical_features] = imputer.transform(electricity_data[numerical_features])
electricity_data.head()
```

```python
electricity_data.describe()
```

```python
sns.catplot(x='Region', y='2014', kind="box",  data=electricity_data, height=6,aspect=2.5)
sns.stripplot(x='Region', y='2014', data=electricity_data, alpha=0.3,jitter=0.2,color='k')
```

```python
sns.catplot(x='IncomeGroup', y='2014', kind="box",  data=electricity_data,height=6,aspect=2.5)
sns.stripplot(x='IncomeGroup', y='2014', data=electricity_data, alpha=0.3,jitter=0.2,color='k')
```
