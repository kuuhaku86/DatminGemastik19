```python
numerical_features = electricity_data.select_dtypes(include='float64')
numerical_features = numerical_features.columns
for i in range(3,14):
    plt.figure(figsize=(35,6))
    plt.plot(numerical_features,electricity_data.loc[i,numerical_features])
    plt.show()
```

```python
electricity_data.describe()
```

```python
print(electricity_data.isnull().sum())
from sklearn.preprocessing import Imputer
electricity_data = electricity_data.drop(['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970'],axis=1)
numerical_features = electricity_data.select_dtypes(include='float64')
numerical_features = numerical_features.columns
imputer = Imputer(missing_values = 'NaN',strategy='median',axis=0)
imputer = imputer.fit(electricity_data[numerical_features])
electricity_data[numerical_features] = imputer.transform(electricity_data[numerical_features])
electricity_data.head()
```

```python
electricity_data = electricity_data[electricity_data['IncomeGroup'].notnull()]
```
