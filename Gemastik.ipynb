```python
print(electricity_data.isnull().sum())
from sklearn.preprocessing import Imputer
electricity_data = electricity_data.drop(['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970'],axis=1)
numerical_features = electricity_data.select_dtypes(include='float64')
numerical_features = numerical_features.columns
imputer = Imputer(missing_values = 'NaN',strategy='median',axis=0)
imputer = imputer.fit(electricity_data[numerical_features])
electricity_data[numerical_features] = imputer.transform(electricity_data[numerical_features])
electricity_data.head()
```

```python
electricity_data = electricity_data[electricity_data['IncomeGroup'].notnull()]
```

```python
import numpy as np
y = electricity_data['2014']
selected_features = np.concatenate((numerical_features,['IncomeGroup']),axis=0)
X = electricity_data[selected_features]
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,random_state=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[['IncomeGroup']]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[['IncomeGroup']]))

OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

num_X_train = X_train.drop(['IncomeGroup'], axis=1)
num_X_valid = X_valid.drop(['IncomeGroup'], axis=1)

X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
print(X_train.head())
print(X_train.shape)
print(X_valid.shape)
print(y_train.head())
print(y_train.shape)
print(y_valid.shape)
```

```python
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.layers import Flatten
from sklearn.metrics import mean_squared_error
```

```python
#vanilla LSTM
X_train, y_train = np.array(X_train), np.array(y_train)
X_train_1 = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_valid, y_valid = np.array(X_valid), np.array(y_valid)
X_valid_1 = np.reshape(X_valid, (X_valid.shape[0], X_valid.shape[1], 1))

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(48, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train_1,y_train,epochs=200, verbose=0)

y_res = model.predict(X_valid_1,verbose=0)
y_res = y_res[:,0]
print('RMSE = ',np.sqrt(mean_squared_error(y_valid,y_res)))
y_res = pd.DataFrame(data=y_res)
y_valid_1 = pd.DataFrame(data=y_valid)
y_res.columns = ['forecast']
y_valid_1.columns = ['actual']
print(y_res.describe(),'\n',y_valid_1.describe())
ax = plt.gca()
y_res.plot(kind='line',y='forecast',ax=ax)
y_valid_1.plot(kind='line',y='actual',ax=ax)
plt.show()
```

```python
#stacked LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(48, 1)))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train_1, y_train, epochs=200, verbose=0)
y_res = model.predict(X_valid_1, verbose=0)
y_res = y_res[:,0]
print('RMSE = ',np.sqrt(mean_squared_error(y_valid,y_res)))
y_res = pd.DataFrame(data=y_res)
y_valid_1 = pd.DataFrame(data=y_valid)
y_res.columns = ['forecast']
y_valid_1.columns = ['actual']
print(y_res.describe(),'\n',y_valid_1.describe())
ax = plt.gca()
y_res.plot(kind='line',y='forecast',ax=ax)
y_valid_1.plot(kind='line',y='actual',ax=ax)
plt.show()
```

```python
#bidirectional LSTM
from keras.layers import Bidirectional
model = Sequential()
model.add(Bidirectional(LSTM(50, activation='relu'), input_shape=(48, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train_1, y_train, epochs=200, verbose=0)
y_res = model.predict(X_valid_1, verbose=0)
y_res = y_res[:,0]
print('RMSE = ',np.sqrt(mean_squared_error(y_valid,y_res)))
y_res = pd.DataFrame(data=y_res)
y_valid_1 = pd.DataFrame(data=y_valid)
y_res.columns = ['forecast']
y_valid_1.columns = ['actual']
print(y_res.describe(),'\n',y_valid_1.describe())
ax = plt.gca()
y_res.plot(kind='line',y='forecast',ax=ax)
y_valid_1.plot(kind='line',y='actual',ax=ax)
plt.show()
```

```python
#vector output model LSTM
model = Sequential()
model.add(LSTM(100, activation='relu', return_sequences=True, input_shape=(48, 1)))
model.add(LSTM(100, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.fit(X_train_1, y_train, epochs=200, verbose=0)
y_res = model.predict(X_valid_1, verbose=0)
y_res = y_res[:,0]
print('RMSE = ',np.sqrt(mean_squared_error(y_valid,y_res)))
y_res = pd.DataFrame(data=y_res)
y_valid_1 = pd.DataFrame(data=y_valid)
y_res.columns = ['forecast']
y_valid_1.columns = ['actual']
print(y_res.describe(),'\n',y_valid_1.describe())
ax = plt.gca()
y_res.plot(kind='line',y='forecast',ax=ax)
y_valid_1.plot(kind='line',y='actual',ax=ax)
plt.show()
```
