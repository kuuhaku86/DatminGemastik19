```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filepath = "data.xls"
electricity_data = pd.read_excel(filepath)
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN',strategy='mean',axis=0)
numerical_features = electricity_data.select_dtypes(include='float64')
numerical_features = numerical_features.columns
imputer = imputer.fit(electricity_data[numerical_features])
electricity_data[numerical_features] = imputer.transform(electricity_data[numerical_features])
print(electricity_data.head())
```

```python
electricity_data.describe()
```

```python
corr = electricity_data["IncomeGroup","Region","2014"].corr()
plt.subplots(figsize=(16,9))
heatmap=sns.heatmap(corr,
                    xticklabels=corr.columns,
                    yticklabels=corr.columns)
```
